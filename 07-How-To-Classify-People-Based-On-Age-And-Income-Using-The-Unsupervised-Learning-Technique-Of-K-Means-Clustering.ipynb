### How To Classify People Based On Age And Income Using The "Unsupervised Learning" Technique of K-Means Clustering

To start out, we will fabricate 3 clusters of "age vs. income" data for 100 people.

```python
%matplotlib inline

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
from numpy import random, array

def createClusteredData(N, k):
    random.seed(10)
    pointsPerCluster = float(N)/k
    X = []
    for i in range (k):
        incomeCentroid = random.uniform(20000.0, 200000.0)
        ageCentroid = random.uniform(20.0, 70.0)
        for j in range(int(pointsPerCluster)):
            X.append([random.normal(incomeCentroid, 10000.0), random.normal(ageCentroid, 2.0)])
    X = array(X)
    return X

data = createClusteredData(100, 3)
```

Next we'll use the unsupervised learning model of "K-Means Clustering" to find these clusters.

Naturally, since we just created our own data set,we already know that "3" is the correct value for "k". However, in reality, when using our "K Means Clustering" algorithm to find clusters, we won't know this number in advance.

So we will have to experiment with some values to hone in on the right value for "k".

Let's start with a "k value" of 2 to see what happens when we try to divide our data into 2 clusters.

```python
# Build a "K-Means Cluster" model with 2 clusters
model = KMeans(n_clusters=2)

# Fit the model with scaled (normalized) data
model.fit(scale(data))

# Visualize the clusters
plt.figure(figsize=(8, 6))
plt.scatter(data[:,0], data[:,1], c=model.labels_.astype(float))
plt.show()
```

Hmmm... looks like 2 is too low of a "k" value.

How about 4?

```python
# Build a "K-Means Cluster" model with 4 clusters
model = KMeans(n_clusters=4)

# Fit the model with scaled (normalized) data
model.fit(scale(data))

# Visualize the clusters
plt.figure(figsize=(8, 6))
plt.scatter(data[:,0], data[:,1], c=model.labels_.astype(float))
plt.show()
```

Looks like 4 is too high of a "k" value.

How about 3?

```python
# Build a "K-Means Cluster" model with 3 clusters
model = KMeans(n_clusters=3)

# Fit the model with scaled (normalized) data
model.fit(scale(data))

# Visualize the clusters
plt.figure(figsize=(8, 6))
plt.scatter(data[:,0], data[:,1], c=model.labels_.astype(float))
plt.show()
```

Based on the visual correspondence of colors to clusters, it definitely appears that our "k value" of 3 is the winner here!
